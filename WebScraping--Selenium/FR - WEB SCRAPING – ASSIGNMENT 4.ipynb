1. Scrape the details of most viewed videos on YouTube from Wikipedia.

    - Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

    You need to find following details:
        - A) Rank
        - B) Name
        - C) Artist
        - D) Upload date
        - E) Views

```python
PATH = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(PATH)
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.maximize_window()
```

```python
Rank=[]
Name = []
Artist = []
Views = []
Upload_Date = []
main1=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")
for i in main1:
    rank = i.text.strip(".")
    Rank.append(rank)
Rank = Rank[:30]
Name = []
main2 = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")
for i in main2:
    name = i.text.split('[')[0]
    Name.append(name)
Name = Name[:30]
main3 = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")
for i in main3:
    artist = i.text
    Artist.append(artist)
Artist = Artist[:30]
main4 = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")
for i in main4:
    views = i.text
    Views.append(views)
Views = Views[:30]
main5 = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")
for i in main5:
    date = i.text
    Upload_Date.append(date)
Upload_Date = Upload_Date[:30]
wiki_most_viewed_vids = pd.DataFrame()
wiki_most_viewed_vids['Rank']=Rank
wiki_most_viewed_vids['Name']=Name
wiki_most_viewed_vids['Artist']=Artist
wiki_most_viewed_vids['Views (Billions)']=Views
wiki_most_viewed_vids['Upload Date']=Upload_Date
driver.close()
wiki_most_viewed_vids
```

```python
Name = []
Designation = []
Company = []
Skills = []
Location = []
Number = int(input('Enter the number of pages to scrap the data from :'))
PATH = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(PATH)
url= 'https://www.naukri.com/hr-recruiters-consultants'

driver.get(url)
driver.maximize_window()
time.sleep(5)

driver.find_element_by_xpath('//*[@id="skill"]/div[1]/div[2]/input').send_keys('Data Science')

driver.find_element_by_id("qsbFormBtn").click()

for i in range(1,Number):
    try:
        main = driver.find_elements_by_xpath('//p[@class="highlightable"]')
        for i in main:
            try:
                Location.append(i.text.split('\n')[3])
                Name.append(i.text.split('\n')[0])
                Designation.append(i.text.split('\n')[1])
                Company.append(i.text.split('\n')[2])
            except:
                Location.append('-')
                Name.append('-')
                Designation.append('-')
                Company.append('-')
        main1 = driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
        for i in main1:
            try:
                Skills.append(i.text)
            except:
                Skills.append('-')
    except:
        pass
    try:
        driver.find_element_by_xpath('//div[@class="pagination"]//button[2]').click()
    except:
        driver.find_element_by_xpath('//div[@class="pagination"]//button[1]').click()
driver.close()
Naukri = pd.DataFrame()
Naukri['Name'] = Name
Naukri['Designation'] = Designation
Naukri['Company'] = Company
Naukri['Skills'] = Skills
Naukri['Location'] = Location
Naukri
```

        - B) Year span
        - C) Genre
        - D) Run time
        - E) Ratings
        - F) Votes

```python
PATH = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(PATH)
driver.get('https://www.imdb.com/list/ls095964455/')
driver.maximize_window()
time.sleep(5)
```

```python
Rank = []
Name = []
Year_Span = []
Genre = []
Run_Time = []
Ratings = []
Votes = []
r = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/span[1]')
for i in r:
    rank = i.text
    Rank.append(rank)
n = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//a')
for i in n:
    name = i.text
    Name.append(name)
y = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//span[2]')
for i in y:
    year = i.text.strip('()')
    Year_Span.append(year)
g = driver.find_elements_by_xpath('//p[@class="text-muted text-small"]//span[5]')
for i in g:
    genre = i.text
    Genre.append(genre)
rt = driver.find_elements_by_xpath('//p[@class="text-muted text-small"]//span[3]')
for i in rt:
    run = i.text
    Run_Time.append(run)
rat = driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]')
for i in rat:
    rating = i.text
    Ratings.append(rating)
v = driver.find_elements_by_xpath('//p[@class="text-muted text-small"]//span[2]')
for i in v:
    vote = i.text
    if vote == '|':
        pass
    else:
        Votes.append(vote)
driver.close()
imdb = pd.DataFrame()
imdb['Rank'] = Rank
imdb['Name'] = Name
imdb['Year Span'] = Year_Span
imdb['Genre'] = Genre
imdb['Run Time'] = Run_Time
imdb['Ratings'] = Ratings
imdb['Votes'] = Votes
imdb = imdb.set_index('Rank')
imdb
```
