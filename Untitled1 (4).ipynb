## 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts . 

```python
from bs4 import BeautifulSoup
import requests
import lxml
import numpy as np
import pandas as pd
```

```python
import os
from selenium import webdriver
PATH = 'C:\Program Files (x86)\chromedriver.exe'
url = 'https://www.bewakoof.com/women-clothing'
driver = webdriver.Chrome(PATH)
driver.get(url)

page = driver.page_source
soup = BeautifulSoup(page)
bewakoof = []
for main_file in soup.find_all('div',class_="productCardDetail"):
        name = main_file.find('h3').text
        price = main_file.find('span',attrs={'class' :"discountedPriceText "})
        actual_price = main_file.find('span',class_="actualPriceText ")
        if (len(name))>0 or len(price)>0:
            bewakoof.append([name,price,actual_price])
df = pd.DataFrame(bewakoof)
df
```
