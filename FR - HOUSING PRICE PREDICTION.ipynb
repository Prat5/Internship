### Handling Missig Values

###### We could drop the columns with over 900 NUll values and Id Columns as it's not of any use in Model.

```python
df.drop(['MiscFeature','Fence','PoolQC','Alley','Id'],axis=1,inplace=True)
df.head()
```

```python
df_test.drop(['MiscFeature','Fence','PoolQC','Alley','Id'],axis=1,inplace=True)
df_test.head()
```

```python
df_test.head(10)
```

##### - LotFrontage

```python
pivot1 = df.pivot_table(values = 'LotFrontage',columns='MSZoning')
pivot1
```

Lets fill the Missing values in LotFrontage based on MSZoning

```python
missing = df['LotFrontage'].isnull()
```

```python
df.loc[missing,'LotFrontage'] = df.loc[missing,'MSZoning'].apply(lambda x: pivot1[x])
```

```python
df.head()
```

```python
missing = df_test['LotFrontage'].isnull()
```

```python
df_test.loc[missing,'LotFrontage'] = df_test.loc[missing,'MSZoning'].apply(lambda x: pivot1[x])
```

```python
df_test.head()
```

```python
df['LotFrontage'].isnull().sum(),df_test['LotFrontage'].isnull().sum()
```

###### - MasVnrArea

let's fill MasVnrArea with mean

```python
df['MasVnrArea'].fillna(df['MasVnrArea'].mean(),inplace=True)
```

```python
df_test['MasVnrArea'].fillna(df_test['MasVnrArea'].mean(),inplace=True)
```

###### - GarageYrBlt

```python
pivot2 = df.pivot_table(values = 'GarageYrBlt',index = 'YearBuilt')
pivot2['GarageYrBlt'] = np.round(pivot2['GarageYrBlt'])
pivot2.tail(10)
```

Filling Garrage year buit with mean would be in appropriate hence filling GarageYrBlt based on YearBuilt as we see a strong relation based on above.

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['YearBuilt'])
```

```python
df_test['GarageYrBlt'] = df_test['GarageYrBlt'].fillna(df['YearBuilt'])
```

###### - Filling Missing Values in remaining categorical columns with Mode

```python
null = []
for i in df.columns:
    if df[i].isnull().sum() > 0: 
        null.append(i)
null
```

```python
for i in null:
    df[i] = df[i].fillna(df[i].mode()[0])
df
```

```python
df.isnull().sum()
```

We do not have any null values any more
